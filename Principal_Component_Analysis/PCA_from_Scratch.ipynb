```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
%matplotlib inline

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
features = ['n1', 'n2', 'n3', 'n4']

experiment_data = load_iris()

x=experiment_data.data
y=experiment_data.target

print(x)
print(experiment_data.target_names)


```

The aim is to standardize the range of the continuous initial variables so that each one of them contributes equally to the analysis.

PCA is quite sensitive regarding the variances of the initial variables.


```python
x_std = StandardScaler().fit_transform(x)  
print(x_std)
```

# Covariance Matrix of Features

The aim of this step is to understand how the variables of the input data set are varying from the mean with respect to each other, or in other words, to see if there is any relationship between them.

```python
#features are columns from x_std
features = x_std.T

covariance_matrix = np.cov(features)
print(covariance_matrix)
```

# Eigen Vectors and Values from Covariance Matrix
Eigenvectors and eigenvalues are the linear algebra concepts that we need to compute from the covariance matrix in order to determine the principal components of the data.

An important thing to realize here is that, the principal components are less interpretable and donâ€™t have any real meaning since they are constructed as linear combinations of the initial variables.

```python
eig_vals, eig_vecs = np.linalg.eig(covariance_matrix)

print(eig_vals)

print(eig_vecs)
```

```python
#print(len(eig_vals))
for i in range(4):
    print(eig_vals[i]/sum(eig_vals))

print(eig_vecs.T[0])
```

# Project data point onto selected Eigen Vectors



```python
projected_x = x_std.dot(eig_vecs.T[0])
projected_x
```

```python
result = pd.DataFrame(projected_x, columns=['PC1'])


result.head(20)
```
